### Second Layer

```python
selector = alt.selection_single(encodings=['longitude','latitude'], nearest=True)

condition = alt.condition(selector, alt.OpacityValue(1), alt.value(0.05))

secondlayer = alt.Chart.from_dict(
{
   "width":800,
   "height":500,
   "projection":{
      "type":"albersUsa"
   },
   "data":{
      "url":"https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_bcubcg_fall2022/main/data/bfro_reports_fall2022.csv"
   },
   "transform":[
      {
         "filter":"(datum.latitude > 55 && datum.latitude < 80 && datum.longitude > -170 && datum.longitude < -140) || (datum.latitude > 25 && datum.latitude < 50 && datum.longitude > -140)"
      }
   ],
   "encoding":{
      "longitude":{
         "field":"longitude",
         "type":"quantitative"
      },
      "latitude":{
         "field":"latitude",
         "type":"quantitative"
      }
   },
   "mark":{
      "type":"circle",
      "color":"red",
      "size":50
   }
}
).add_selection(selector)

secondlayer.encoding.opacity = condition

secondlayer
```

### First Layer

```python
firstlayer = alt.Chart.from_dict(
{
   "data": {
        "url": "https://raw.githubusercontent.com/vega/vega/main/docs/data/us-10m.json",
        "format": {
          "type": "topojson",
          "feature": "states"
        }
      },
      "mark": {
        "type": "geoshape",
        "fill": "lightgray",
        "stroke": "white"
      }
}
)
firstlayer
```
